```python
# تعريف وادراج مكتبة بانداز
import pandas as pd
```

```python
# قراءة البيانات من الملف
mobile = pd.read_csv("/content/test.csv")
mobile
```

```python

# الاحصائيات الرقمية
mobile.describe()
print(mobile)
# عرض ملخص عن البيانات
print("أسماء الأعمدة في البيانات:")
print(mobile.columns)

print("\nعدد الصفوف والأعمدة:")
print(mobile.shape)

print("\nمعاينة أول 5 صفوف:")
print(mobile.head())

print("\nعدد القيم الفارغة في كل عمود:")
print(mobile.isnull().sum())

print("\nمعلومات عامة عن البيانات:")
print(mobile.info())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
#  يمكننا استخدام Seaborn لرسم توزيع القيم في عمود معين مثل "battery_power".

plt.figure(figsize=(10, 6))
sns.histplot(mobile['battery_power'], kde=True, color='skyblue')

plt.title("battery_power")
plt.xlabel("Capacity_battery")
plt.ylabel("number of mobile")
plt.show()
```

```python
#  العلاقة بين سعة البطارية و مدة المكالمات

plt.figure(figsize=(10, 6))
sns.scatterplot(x='battery_power', y='talk_time', data=mobile, color='green')

plt.title("The relationship between battery capacity and call duration")
plt.xlabel("battery capacity")
plt.ylabel("call duration in hours")
plt.show()
```

```python
#  عرض البيانات بطريقة عمودية سهلة الفهم، مثل عدد الهواتف التي تدعم 3G و4G

three_g_count = mobile['three_g'].value_counts()
four_g_count = mobile['four_g'].value_counts()

plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
three_g_count.plot(kind='bar', color=['orange', 'blue'])
plt.title('number of mobile supports 3G')
plt.xlabel('Supports 3G')
plt.ylabel('numper of mobile')

plt.subplot(1, 2, 2)
four_g_count.plot(kind='bar', color=['green', 'red'])
plt.title('number of mobile supports 4G')
plt.xlabel('Supports 4G')
plt.ylabel(' numper of mobile')

plt.tight_layout()
plt.show()
```

```python
# حساب نسب الهواتف التي تدعم Wi-Fi

wifi_counts = mobile['wifi'].value_counts()

plt.figure(figsize=(7, 7))
plt.pie(wifi_counts, labels=['Support Wi-Fi', ' not support Wi-Fi'], autopct='%1.1f%%', colors=['lightblue', 'lightcoral'])
plt.title('Percentage phone that supports Wi-Fi')
plt.show()
```

```python
#  عرض نسبة الهواتف التي تدعم Wi-Fi مقابل التي لا تدعمه.

wifi_counts = mobile['wifi'].value_counts()

plt.figure(figsize=(7, 7))
plt.pie(wifi_counts, labels=['Support Wi-Fi', 'not support Wi-Fi'], autopct='%1.1f%%', colors=['lightblue', 'lightcoral'])
plt.title(' Percentage phone that supports Wi-Fi')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='battery_power', y='talk_time', data=mobile.sample(20), palette='viridis')

plt.title("The relationship between battery capacity and call duration")
plt.xlabel("battery capacity (mAh)")
plt.ylabel("call duration in hours")
plt.xticks(rotation=45)
plt.show()
```

```python
#

mobile['battery_category'] = pd.cut(mobile['battery_power'], bins=[0, 1000, 2000, 3000, 4000, 5000], labels=['<1000', '1000-2000', '2000-3000', '3000-4000', '>4000'])

battery_distribution = mobile['battery_category'].value_counts()

plt.figure(figsize=(8, 8))
plt.pie(battery_distribution, labels=battery_distribution.index, autopct='%1.1f%%', colors=sns.color_palette("coolwarm"))

plt.title("battery capacity distribution ratio")
plt.show()
```

```python
#  يمكن استخدام Bar Plot للمقارنة بين البطارية ومدة المكالمات عبر تقسيم القيم.

plt.figure(figsize=(12, 6))
sns.barplot(x='battery_power', y='talk_time', data=Mobile.sample(20), palette='viridis')

plt.title("Relationship between Battery Capacity and Talk Time")
plt.xlabel("Battery Capacity (mAh)")
plt.ylabel("Talk Time (hours)")
plt.xticks(rotation=45)
plt.show()
```

```python
#  Histogram نرسم رسمين بيانيين لتوضيح توزيع دقة الكاميرا الأمامية والخلفية باستخدام .

plt.subplot(1, 2, 1)
sns.histplot(mobile['fc'], kde=True, color='blue')
plt.title(' Front camera distribution accuracy (fc)')
plt.xlabel(' Front camera accuracy (MP)')

plt.subplot(1, 2, 2)
sns.histplot(mobile['pc'], kde=True, color='green')
plt.title(' rear camera distribution accuracy (pc)')
plt.xlabel(' rear camera accuracy (MP)')

plt.tight_layout()
plt.show()
```

```python
#  لتوضيح العلاقة بين ارتفاع وعرض الشاشة بالبيكسل.

plt.figure(figsize=(10, 6))
sns.scatterplot(x='px_width', y='px_height', data=mobile, color='purple', alpha=0.6)

plt.title('the relationship bettween screen width and heigth')
plt.xlabel('px_width')
plt.ylabel('px_height')
plt.show()
```

```python
#  مقارنة عدد الهواتف التي   3G، 4G، و Wi-Fi

plt.figure(figsize=(12, 5))

# support 3G
plt.subplot(1, 3, 1)
sns.countplot(x='three_g', data=mobile, palette='Set2')
plt.title('support 3G')
plt.xlabel('support 3G ( 1 = Yes , 0 = no )')
plt.ylabel('number of phone')

# support 4G
plt.subplot(1, 3, 2)
sns.countplot(x='four_g', data=mobile, palette='Set1')
plt.title('support 4G')
plt.xlabel('support 4G ( 1 = Yes , 0 = no )')

# support Wi-Fi
plt.subplot(1, 3, 3)
sns.countplot(x='wifi', data=mobile, palette='coolwarm')
plt.title('support Wi-Fi')
plt.xlabel('support Wi-Fi ( 1 = Yes , 0 = no )')

plt.tight_layout()
plt.show()
```

```python
# لرسم مخطط خطي أو منحنى يوضح العلاقة بين وزن الهاتف (mobile_wt) و سعة البطارية (battery_power)،

sorted_data = mobile.sort_values(by='mobile_wt')

plt.figure(figsize=(10, 6))
plt.plot(sorted_data['mobile_wt'], sorted_data['battery_power'], color='blue', marker='o', linestyle='-', alpha=0.7)

plt.title("Relationship between Mobile Weight and Battery Capacity", fontsize=14)
plt.xlabel("Mobile Weight (grams)", fontsize=12)
plt.ylabel("Battery Capacity (mAh)", fontsize=12)
plt.grid(alpha=0.3)
plt.show()
```

```python


plt.figure(figsize=(10, 6))
sns.regplot(x='mobile_wt', y='battery_power', data=mobile, scatter_kws={'alpha':0.6}, line_kws={"color": "red"})

plt.title("Relationship between Mobile Weight and Battery Capacity (Curve)", fontsize=14)
plt.xlabel("Mobile Weight (grams)", fontsize=12)
plt.ylabel("Battery Capacity (mAh)", fontsize=12)
plt.grid(alpha=0.3)
plt.show()
```

```python
#العلاقة بين دقة الشاشة وسعة  RAM أو سعة البطارية لمعرفة ما إذا كانت الشاشات عالية الدقة تحتاج إلى موارد أكثر.

plt.figure(figsize=(10, 6))
sns.scatterplot(x='px_width', y='px_height', size='ram', hue='ram', data=mobile, alpha=0.6, palette='cool')

plt.title("Relationship between Screen Resolution and RAM")
plt.xlabel("Screen Width (px)")
plt.ylabel("Screen Height (px)")
plt.colorbar(label="RAM (MB)")
plt.show()
```

```python
#العلاقة بين عدد أنوية المعالج وسعة البطارية أو مدة المكالمات لمعرفة تأثير أداء المعالج على عمر البطارية.

plt.figure(figsize=(10, 6))
sns.barplot(x='n_cores', y='battery_power', data=mobile, palette='viridis')

plt.title("Battery Capacity by Number of Processor Cores")
plt.xlabel("Number of Processor Cores")
plt.ylabel("Battery Capacity (mAh)")
plt.show()
```

```python
# هل الهواتف التي تحتوي على ذاكرة داخلية أكبر تحتوي أيضًا على كاميرات خلفية بدقة أعلى؟

plt.figure(figsize=(10, 6))
sns.scatterplot(x='int_memory', y='pc', data=mobile, hue='pc', palette='coolwarm', alpha=0.6)

plt.title("Relationship between Internal Memory and Primary Camera Quality")
plt.xlabel("Internal Memory (GB)")
plt.ylabel("Primary Camera (MP)")
plt.show()
```

```python
# هل الهواتف الأثقل تحتوي على ميزات إضافية مثل دعم الجيل الرابع أو سعة بطارية أكبر؟

plt.figure(figsize=(10, 6))
sns.scatterplot(x='mobile_wt', y='battery_power', hue='four_g', data=mobile, palette='coolwarm', alpha=0.6)

plt.title("Relationship between Mobile Weight and Battery Capacity with 4G Support")
plt.xlabel("Mobile Weight (grams)")
plt.ylabel("Battery Capacity (mAh)")
plt.show()
```

```python
# هل الهواتف ذات سعة  أعلى  RAM تحتوي على بطاريات أكبر لدعم الأداء؟

plt.figure(figsize=(10, 6))
sns.scatterplot(x='ram', y='battery_power', data=mobile, color='purple', alpha=0.6)

plt.title("Relationship between RAM and Battery Capacity")
plt.xlabel("RAM (MB)")
plt.ylabel("Battery Capacity (mAh)")
plt.show()
```

```python
#  خريطة حرارية للعلاقات بين الأعمدة العددية

plt.figure(figsize=(12, 8))
sns.heatmap(mobile.corr(), annot=True, cmap='coolwarm', fmt='.2f')

plt.title('Heat Map')
plt.show()
```

```python
# إضافة خط يوضح العلاقة بين سعة البطارية و مدة المكالمات مع تضمين البيانات النقطية

sorted_data = mobile.sort_values(by='battery_power')

plt.figure(figsize=(12, 6))

# رسم البيانات النقطية
plt.scatter(sorted_data['battery_power'], sorted_data['talk_time'], color='blue', alpha=0.6, label='Data Points')

# رسم الخط
plt.plot(sorted_data['battery_power'], sorted_data['talk_time'], color='orange', label='Trend Line')

# إعداد العناوين والمحاور
plt.title("Relationship between Battery Capacity and Talk Time")
plt.xlabel("Battery Capacity (mAh)")
plt.ylabel("Talk Time (hours)")
plt.legend()
plt.grid(alpha=0.3)
plt.show()
```

```python
#  توضيح العلاقة بين سعة البطارية و مدة المكالمات مع تمييز الهواتف حسب دعم 4G.

plt.figure(figsize=(12, 6))
sns.scatterplot(
    x='battery_power',
    y='talk_time',
    hue='four_g',
    data=mobile,
    palette='coolwarm',
    alpha=0.7
)

# إعداد العناوين والمحاور
plt.title("Battery Capacity vs Talk Time with 4G Support")
plt.xlabel("Battery Capacity (mAh)")
plt.ylabel("Talk Time (hours)")
plt.legend(title="4G Support (1 = Yes, 0 = No)")
plt.grid(alpha=0.3)
plt.show()
```

```python


plt.figure(figsize=(12, 6))

# رسم الفقاعات
plt.scatter(
    mobile['battery_power'],
    mobile['talk_time'],
    s=mobile['ram']/10,
    alpha=0.6,
    c=mobile['ram'],
    cmap='viridis'
)

# إعداد العناوين والمحاور
plt.title("Battery Capacity vs Talk Time (Bubble Plot with RAM)")
plt.xlabel("Battery Capacity (mAh)")
plt.ylabel("Talk Time (hours)")
plt.colorbar(label="RAM (MB)")
plt.grid(alpha=0.3)
plt.show()
```

```python
#  تفاصيل الهواتف (مثل دعم 3G) في الرسم العمودي.

plt.figure(figsize=(14, 6))
sns.barplot(
    x='battery_power',
    y='talk_time',
    hue='three_g',
    data=mobile.sample(20),
    palette='Set2'
)

# إعداد العناوين والمحاور
plt.title("Battery Capacity vs Talk Time (Grouped by 3G Support)")
plt.xlabel("Battery Capacity (mAh)")
plt.ylabel("Talk Time (hours)")
plt.legend(title="3G Support (1 = Yes, 0 = No)")
plt.xticks(rotation=45)
plt.show()
```

```python
#  إظهار العلاقة بين سعة البطارية و مدة المكالمات باستخدام خريطة حرارية.

heatmap_data = mobile.pivot_table(
    index='talk_time',
    columns='battery_power',
    values='ram',
    aggfunc='mean'
)

plt.figure(figsize=(14, 6))
sns.heatmap(heatmap_data, cmap='YlGnBu', cbar_kws={'label': 'Average RAM (MB)'})

# إعداد العناوين والمحاور
plt.title("Heatmap of Battery Capacity vs Talk Time")
plt.xlabel("Battery Capacity (mAh)")
plt.ylabel("Talk Time (hours)")
plt.show()
```
